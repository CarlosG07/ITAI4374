- **1. How does the input current affect the generation of action potentials?** Input current controls whether the neuron will fire or not. When the current is strong enough, it pushes the membrane potential up to a certain level called the threshold. Once that level is reached, the neuron generates an action potential (a spike). In the lab, increasing the input current made the neuron reach the threshold faster, which caused it to fire. So basically, stronger input current makes it easier and quicker for the neuron to produce spikes.

- **2. What happens when the input current is below the threshold?** When the input current is below the threshold, the neuron does not fire an action potential. Instead, the membrane potential only changes a little and then goes back to normal. In the lab, this looked like small bumps or waves instead of spikes. These are called subthreshold responses. Since the threshold is not reached, no signal is sent. This shows that a minimum amount of input is required for a neuron to actually activate.

- **3. How does the neuron's response change with varying input current strengths?** As the input current increases, the neuron’s response changes in how often it fires. With weak current, there are no spikes. Once the threshold is reached, the neuron starts firing. If the current gets even stronger, the neuron fires more frequently, meaning more spikes over time. However, the size of each spike stays about the same. In the lab, this showed that neurons don’t make bigger signals with stronger input—they send signals more often instead.

```python
import numpy as np

class ArtificialNeuron:
    """Implements artificial neuron for comparison"""
    def __init__(self, num_inputs):
        self.weights = np.random.randn(num_inputs)
        self.bias = 0
```

```python
   def activate(self, inputs):
        """Implements artificial neuron activation"""
        # Calculate weighted sum of inputs
        weighted_sum = np.dot(self.weights, inputs) + self.bias
        # Apply activation function (example: sigmoid)
        output = 1 / (1 + np.exp(-weighted_sum))
        return output
```

```python
class ArtificialNeuron:
    def __init__(self, num_inputs):
        self.weights = np.random.rand(num_inputs)
        self.bias = np.random.rand()

    def activate(self, inputs):
        # Weighted sum
        z = np.dot(self.weights, inputs) + self.bias

        # Sigmoid activation function
        output = 1 / (1 + np.exp(-z))
        return output
```

- **1. How do the input-output relationships differ between the two models?**
In the biological neuron, the input-output relationship is time-based and depends on reaching a threshold. Inputs build up over time, and once the threshold is reached, the neuron fires spikes. In the artificial neuron, the output is calculated instantly using a formula (weighted sum + activation function). It gives a direct output value instead of spikes. So, biological neurons produce signals over time, while artificial neurons give immediate numerical outputs based on inputs.

- **2. What are the key distinctions in information processing between biological and artificial neurons?**
Biological neurons process information using electrical signals that change over time and communicate through spikes. They also include more complex behaviors like delays and membrane dynamics. Artificial neurons are much simpler—they just multiply inputs by weights, add them up, and pass the result through an activation function. In your lab, the biological model showed realistic behavior, while the artificial neuron focused on fast, simplified calculations used in machine learning.

- **3. How do the concepts of threshold and activation function relate to each other?**
The threshold in a biological neuron is the point that must be reached for the neuron to fire. In artificial neurons, the activation function plays a similar role by deciding the output based on the input value. For example, functions like ReLU or sigmoid control when and how strongly a neuron “activates.” So, the threshold is like a trigger in biological neurons, while the activation function is a mathematical way to control that behavior in artificial neurons.


```python
class ArtificialNeuron:
    def __init__(self, num_inputs, activation="sigmoid"):
        self.weights = np.random.rand(num_inputs)
        self.bias = np.random.rand()
        self.activation = activation

    def activate(self, inputs):
        # Step 1: Weighted sum
        weighted_sum = np.dot(self.weights, inputs) + self.bias

        # Step 2: Apply activation function
        if self.activation == "sigmoid":
            output = 1 / (1 + np.exp(-weighted_sum))

        elif self.activation == "relu":
            output = max(0, weighted_sum)

        elif self.activation == "tanh":
            output = np.tanh(weighted_sum)

        else:
            raise ValueError("Unknown activation function")

        return output
```

```python
inputs = np.array([0.5, 0.2, 0.8])

# Sigmoid
neuron1 = ArtificialNeuron(3, activation="sigmoid")
print("Sigmoid Output:", neuron1.activate(inputs))

# ReLU
neuron2 = ArtificialNeuron(3, activation="relu")
print("ReLU Output:", neuron2.activate(inputs))

# Tanh
neuron3 = ArtificialNeuron(3, activation="tanh")
print("Tanh Output:", neuron3.activate(inputs))
```
